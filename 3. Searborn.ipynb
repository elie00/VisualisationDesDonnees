## Import des librairies

```python
# Importation des librairies pour la manipulation des donnees
import numpy as np
import  pandas as pd

# Importation des librairies de visualisation
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

## Premier graphique

```python
help(sns.lineplot())
```

```python
x = np.linspace(0, 2 * np.pi, 200)
y = np.sinc(x)

# Créez un DataFrame pandas à partir des tableaux numpy
df = pd.DataFrame({'x': x, 'y': y})

# Utilisez seaborn pour tracer le graphique linéaire en spécifiant les noms de colonnes
sns.lineplot(data=df, x='x', y='y')
```

## Import des données

```python
CheminFichier = r'C:\Users\giant\Downloads\python-visualisation-3000779-main\python-visualisation-3000779-main\01_Visualisation_Données\Data\Dataset_Employes.xlsx'
Dataset_Employes = pd.read_excel(CheminFichier)
Dataset_Employes.head()
```

## Les courbes

### Relation continue simple

```python
sns.lineplot(x = 'Age', y = 'Niveau de satisfaction', data = Dataset_Employes)
```

## Relation continue multiple

```python
sns.relplot(x = 'Age', y = 'Niveau de satisfaction', hue = 'Type Contrat', data= Dataset_Employes, kind= 'line', ci = None)
```

# Diagrammes

```python
Dataset_Employes['Service'].value_counts()
```

## Diagramme en barres simples

```python
sns.countplot(x = 'Service', data = Dataset_Employes, palette= 'pastel' )
```

```python
sns.set(rc={'figure.figsize':(10,3)})
sns.countplot(x = 'Service', data = Dataset_Employes, palette= 'pastel')
```

## Diagramme en barres regroupées

```python
sns.catplot(x="Service", y="Salaire base mensuel", hue="Contrat",
            data=Dataset_Employes, kind="bar", ci=None, palette= sns.color_palette("Set2"))
```

# Histogramme

```python
sns.displot(Dataset_Employes['Niveau de satisfaction'], kde=True)

# KDE : Kernel Density Estimator (KDE)
```

```python
# Chosir le nombre de barres

sns.displot(Dataset_Employes['Niveau de satisfaction'], kde=True, bins=10)
```

# Boxplot (Diagramme à moustaches)

```python
Dataset_Employes.describe()
```

```python
sns.boxplot(Dataset_Employes['Age'])
```

```python
sns.boxplot(x = Dataset_Employes['Age'], y = Dataset_Employes['Service'], palette= 'husl')
```

```python
sns.set(style = "whitegrid", palette = "pastel")
sns.boxplot(x = 'Age', y = 'Service', hue = 'Type Contrat', data = Dataset_Employes, palette=["m", "g"])
```
